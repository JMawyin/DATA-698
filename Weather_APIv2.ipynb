```python
def flatten_json(nested_json, exclude=['']):
    """Flatten json object with nested keys into a single level.
        Args:
            nested_json: A nested json object.
            exclude: Keys to exclude from output.
        Returns:
            The flattened json object if successful, None otherwise.
    """
    out = {}

    def flatten(x, name='', exclude=exclude):
        if type(x) is dict:
            for a in x:
                if a not in exclude: flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(nested_json)
    return out

test_json_loc = 'http://datapoint.metoffice.gov.uk/public/data/txt/wxfcs/regionalforecast/json/sitelist?key={}'.format(MET_API_K)
test_pd = pd.read_json(test_json_loc, orient='records')
test_pd.head()

#daily_gen = pd.DataFrame([flatten_json(x) for x in test_pd['data']])
daily_gen = pd.DataFrame([flatten_json(x) for x in test_pd['Locations']['Location']])
#daily_gen.shape
daily_gen.head(10)
```
